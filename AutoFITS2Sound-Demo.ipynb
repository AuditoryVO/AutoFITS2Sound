## AutoFITS2Sound
Automatic auditory exploration of stellar spectrum and light curve catalogs stored in FITS files. Data: SVO.

Data download: http://svocats.cab.inta-csic.es/stelib/index.php?action=search
#### ACKNOWLEDGMENT
This work is based on data from the STELIB service developed by the Spanish Virtual Observatory in the framework of the IAU Comission G5 Working Group : Spectral Stellar Libraries.

### Running a sequential auditory representation of the library

```python
#All in a loop
curves=0
for path, subdirs, files in os.walk(root):
    for name in files:
        curves=curves+1
        file = [os.path.join(path, name)]
        str = " " 
        Ffile = (str.join(file))
        route = Path(name)
        Fname = route.with_suffix('')
        Fpng = route.with_suffix('.png')

        #Plotting each spectrum and saving the image for the multimodal representation in Cabbage 
        data, header = fits.getdata(Ffile, header=True)
        hdu_number = 0
        fits.getheader(Ffile, hdu_number)
        fits_file = Ffile
        
        #Opening FITS file and normalizing the flux
        with fits.open(fits_file, mode='readonly') as hdulist:
            hdulist.info()
            data = hdulist[0].data         
            data_norm = data/(np.nanmax(data))
            
            #Calculating the wavelength axis to represent the spectrum
            wave = np.ones(hdulist[0].header['NAXIS1'], dtype=float)
            for i in range(hdulist[0].header['NAXIS1']):
                wave[i] = (hdulist[0].header['CRVAL1'] + i*hdulist[0].header['CDELT1'])
        
        #Plotting the spectrum        
        fig, ax = plt.subplots()
        fig.set_size_inches(12., 8.)
        fig.suptitle(fits_file)
        plt.plot(wave, data_norm)
        ax.set_ylabel("ADU")
        ax.set_xlabel("wavelength (Å)")
        #Sequential overwriting Spectra.png
        plt.savefig('Spectra.png')
        
        #Initializing and sending OSC messages        
        client1 = udp_client.SimpleUDPClient("127.0.0.1", 9999) #x
        client2 = udp_client.SimpleUDPClient("127.0.0.1", 9998) #y
        client3 = udp_client.SimpleUDPClient("127.0.0.1", 9997) #length
        client4 = udp_client.SimpleUDPClient("127.0.0.1", 9996) #s: Image change
        client5 = udp_client.SimpleUDPClient("127.0.0.1", 9995) #index
        client6 = udp_client.SimpleUDPClient("127.0.0.1", 9994) #play/stop
        
        length = len(data)

        #User definable range for the graphic representation:
        t2= np.nanmax(wave) #Horizontal axis Right limit          
        t1= np.nanmin(wave) #Horizontal axis Left limit           
        max = np.nanmax(data)+20 #Vertical axis Max.limit   
        min = np.nanmin(data)-20 #Vertical axis Min.limit 

        #User definable range for the auditory representation:
        i2= length #Equivalent index for "t2"                        
        i1= 0 #Equivalent index for "t1"                             

        #Speed control for the auditory representation:
        s=1000    
        
        #Sending OSC messages
        client4.send_message("/s", s)
        client4.send_message("/s", 1)
        client6.send_message("/p", 1)

        for i in range (i1, i2):
            print("index:", i)
            print("Wavelength (Angstrom):", wave[i]);
            print("Flux (ADU):", data_norm[i]);
        
            x=float(wave[i])
            y=float(data_norm[i])
    
            client5.send_message("/i", i)
            client1.send_message("/x", x)   
            client2.send_message("/y", y)
            time.sleep(1/s)
    
        client1.send_message("/x", 0)
        client2.send_message("/y", 0)
        client6.send_message("/p", 0)
        client4.send_message("/s", 0)
        plt.close(fig)

        print ("Spectra represented:",curves, "curves");
```
